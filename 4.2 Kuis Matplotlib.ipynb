```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Soal 1
Dari data x dan y buatlah sebuah figure dengan ketentuan :
1. Judul figure adalah Data Penjualan PT. Sejahtera
2. X label adalah Pembelian
3. Y label adalah penjualan

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x, y, 'b')
ax.set_xlabel('Pembelian')
ax.set_ylabel('Penjualan')
ax.set_title('Data penjualan PT. Sejahtera')
```

## Soal 2

Buatlah sebuah figure dengan ketentuan :

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
```

sekarang isi data dari masing - masing axes tersebut dengan x dan y

```python
fig = plt.figure()

ax1 = fig.add_axes([0,0,1,1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

#pengaturan plot utama
ax1.plot(x, y, 'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')

#pengaturan plot dalam
ax2.plot(x, y, 'b')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
```
